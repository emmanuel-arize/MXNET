# Linear Regression


In machine learning, the dataset is called a training data or training set, and the rows are called instances (observations or examples) and what we want to predict is called a target or label or dependent variable. The variables upon which the predictions are based are called features or covariates or independent variables (any column value in the input dataset).


# Linear Model
using age and area of a house as features in predictint the price of a house the linearity assumption  says that the target (price) can be expressed as a weighted sum of the features (area and age):

$$ price = w_{1} *area + w_{2} * age + b$$

where $w_{1}$ and $w_{2}$ are weights associted with area and age respectively and determine the influence of each feature on our prediction, and $b$ is called a bias (also called an offset or intercept)

When our inputs consist of
d features, we express our prediction $\hat y$ as
$\hat y = w_{1}·x_{1} +... + w_{d} · x_{d} + b$
Collecting all features into a vector $x$ and all weights into a vector w, we can express our model
compactly using a dot product:
$$\hat y = w^{T} X + b=Xw + b$$ or $$\hat y = w^{T} X$$ here the $$w_{o}= bias$$

here $X$ is the data points and $w$ is the weight vector


# Loss Function

The loss function quantifies the distance between the actual or real and predicted value of the target(y) with a goal to minimize the prediction error for data point not used in training

# sum of squared errors

the sum of squared errors is the most popular loss function in regression problems and is given by

$$\ell_{i}(w,b)=\frac{1}{2}(y_{i}-\hat y_{i})^{2}$$

 The fraction 1⁄2 is arbitrary, and is used for mathematical
convenience

in general to measure the quality of a model on the entire dataset, we simply average the losses on the training set
$$\ell(w,b)=\frac{1}{2n} \sum_{1}^{n}(w^{T} X_{i}+b -y_{i})^{2}$$

during training we want to find parameters $(w∗,b∗)$ that minimize the total loss across all training samples:
$$w∗, b∗ = argmin_{w,b}L(w, b)$$


# Analytic Solution
Linear regression happens to be a simple optimization problem

$$\ell(w)=(w^{T} X -y_{i})^{2}$$
$$\ell(w)=(w X -y)^{T}(w^{T} X -y)$$

$$\ell=ww^{T}x_{T}x-2w^{T}xy+yy^{T} $$

$$ \frac{\partial \ell}{\partial  w} =2wx^{T}x-2xy=0$$
$$w=(x^{T}x)^{-1}xy$$



# UPDATING PARAMETERS
during trainiing we want to automatically update the parameters

# Learning with the Delta Rule
suppose $$\hat y=u=w_{1}x$$

then the error is given as
$$ E=y- \hat y$$

and the square error $\epsilon$ for an input pattern is
$$ \epsilon =\frac{1}{2}E^{2}=\frac{1}{2}(y- \hat y)^{2} $$


To find the gradient at a particular weight we differentiate the square error function with respect to that particular weight w1, yielding
$$\frac{d \epsilon}{dw_{1}} =\frac{2}{2}(y- w_{1}x)(-x)=-Ex $$


According to the delta rule, weight change is proportional to the negative of the error gradient because it is necessary
to go down the error curve to minimize the error; therefore, the weight change is

$$ \Delta w_{1}\propto Ex$$

where $\propto$ denotes proportionality, which can be replaced with learning rate $\beta$, and the eqution becomes
$$ \Delta w_{1}=\beta Ex$$


The learning rate determines how far along the gradient it is necessary to move to fix the new weights. Larger values of b accelerate the weight change and smaller values slow it down. The new weight after the ith iteration can
be expressed as

$$ W_{1}^{i+1}=W_{1}^{i}- \Delta w_{1}=W_{1}^{i}-\beta Ex $$

using the same idea we can extend to
$$ W_{j}^{i+1}=W_{j}^{i}-\Delta w_{j}^{i}=W_{j}^{i}-\beta Ex_{j}$$ 
$$ W_{0}^{i+1}=W_{0}^{i}-\beta E$$
where $W_{0}$ is the bias

# batch learning
In many situations, it is preferable to wait until all the input patterns (or some portion of them) have been processed and then
adjust weights in an average sense. A batch can be the entire training dataset (epoch-based learning) or some portion of the dataset and is called stochastic gradient descent

$$ \Delta w_{1}=\beta (\frac{1}{n} \sum^{n}Ex_{i})  $$
where n is the batch size

# Normal Distribution
---


$$  p(z)=\frac{1}{ \sqrt[]{2 \pi \sigma^2}} \exp(-\frac{1}{2 \sigma^2}(z-\mu )^{2} )$$



````python
import math
import numpy as np
import d2l
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
def normal(z,mu,sigma):
    p=1/math.sqrt(2*math.pi * sigma**2)
    q=-0.5/ sigma**2
    return p*np.exp( q*(z-mu)**2)
    
````

````python
x = np.arange(-7,7, 0.01)
b=normal(x,mu=0,sigma=1)
x
````

````python
parameters = [(0, 1), (0, 2), (3, 1)]
for mu, sigma in parameters:
    plt.plot(x,normal(x,mu,sigma),label=[mu for mu in parameters])
````

````python
parameters = [(0, 1), (0, 2), (3, 1),(2,4),(0,5)]
d2l.plot(x, [normal(x, mu, sigma) for mu, sigma in parameters], xlabel='z',ylabel='p(z)', 
         figsize=(4.5, 2.5),legend=['mean %d, var %d' % (mu, sigma) for mu, sigma in parameters])
````

from the diagram above one can see that changing the mean corresponds to a shift along the x axis, and increasing the
variance spreads the distribution out, lowering its peak.

# Linear Regression Implementation from Scratch

````python
import d2l
from mxnet import np,npx,autograd
import random
%matplotlib inline
````

# Generating the Dataset

the synthetic dataset generated will be an object $ X^{2}\in R^{1000×2}.$
where 1000 is the number of examples (observations) and 2 is the number of features

````python
# Saved in the d2l package for later use
def synthetic_data(w, b, num_examples):
    """Generate y = X w + b + noise."""
    # np.random.normal(mean,std,size=(rows,columns))
    X = np.random.normal(0, 1, (num_examples, len(w)))
    y = np.dot(X, w) + b
    # y=y+np.random.normal(0, 0.01, y.shape)
    y += np.random.normal(0, 0.01, y.shape)
    return X, y
true_w = np.array([2, -3.4])
true_b = 4.2
features, labels = synthetic_data(true_w, true_b, 1000)
````

````python
print('the data contain '+ str(features.shape[0])+' observations with ' + 
      str(features.shape[1]) +' features(dimension) \n hence a '+ str(features.shape[0]) 
      +" by " + str(features.shape[1])+' matrix' )
````

````python
#first 3 observations
features[:3]
````

````python
# first 2 observations within the target variable
labels[:2]
````

````python
# correlation between the first feature and the target variable(labels) which shows a positive correlation
d2l.set_figsize((3.5, 2.5))
d2l.plt.scatter(features[:, 0].asnumpy(), labels.asnumpy(), 1);
````

````python
# # correlation between the second feature and the target variable(labels) which shows a negative correlation
d2l.set_figsize((3.5, 3.5))
d2l.plt.scatter(features[:, 1].asnumpy(), labels.asnumpy(), 1);
````

````python
def data_iter(batch_size, features, labels):
    num_examples=len(features)# features.shape[0]
    indices=list(range(num_examples))
    # The examples are read at random, in no particular order
    random.shuffle(indices)
    for i in range(0,num_examples,batch_size):
        batch_indices=np.array(indices[i:min(i+batch_size,num_examples)])
        yield features[batch_indices],labels[batch_indices]
````

````python
# letʼs read and print the first small batch of data examples
batch_size = 10
for X, y in data_iter(batch_size, features, labels):
    print(X, '\n \n', y)
    break
````

# Initializing Model Parameters

 initializing the weights by sampling random numbers from a normal distribution with mean 0 and a standard deviation of 0:01,
setting the bias b to 0.

````python
w = np.random.normal(0, 0.01, (2, 1))
b = np.zeros(1)
````

Now that we have initialized our parameters, our next task is to update the various weight and biases until they fit our data
sufficiently well so as to minimize the error.

updating parameters requires taking the gradient (a multi-dimensional derivative) of our loss function with respect to the parameters. Given this gradient, we can update each parameter in the direction that reduces the loss.

````python
# We allocate memory for an ndarrayʼs gradient by invoking its attach_grad method.
w.attach_grad()
b.attach_grad()
````

the calculated gradient taken with respect to x, can be access via the grad attribute with default initialized as an array containing all zeros.

````python
w.grad,b.grad
````

# Defining the Model

````python
def linreg(X, w, b):
    return np.dot(X, w) + b
````

# Defining the Loss Function
updating our model requires taking the gradient of our loss function, so we need to define 
 squared loss function as loss our function for our model. we need to transform the true value y into the predicted valueʼs
shape y_hat

````python
def squared_loss(y_hat, y):
    return 0.5*(y_hat - y.reshape(y_hat.shape)) ** 2 
````

# Defining the Optimization Algorithm
---

# batch learning
updating the weights we define
$$ W_{j}^{i+1}=W_{j}^{i}-\Delta w_{j}^{i}=W_{j}^{i}-\beta Ex_{j}$$ 
$$ W_{0}^{i+1}=W_{0}^{i}-\beta E$$
where $W_{0}$ is the bias


In many situations, it is preferable to wait until all the input patterns (or some portion of them) have been processed and then
adjust weights in an average sense. A batch can be the entire training dataset (epoch-based learning) or some portion of the dataset and is called stochastic gradient descent

$$ \Delta w_{j}=\beta (\frac{1}{n} \sum^{n}Ex_{j})  $$
where n is the batch size

````python
#  stochastic gradient descent (SGD)
def sgd(params, lr, batch_size):
    for para in params:
        para[:]=para-lr*para.grad/batch_size
````

````python
lr = 0.01 # Learning rate
num_epochs = 10 # Number of iterations
lreg = linreg # Our fancy linear model
loss = squared_loss # 0.5 (y-y')^2

for epoch in range(num_epochs):
    for X,y in data_iter(batch_size,features,labels):
        with autograd.record():
            l=loss(lreg(X,w,b),y)# Minibatch loss in X and y
        l.backward() # Compute gradient on l with respect to [w, b]
        sgd([w, b], lr, batch_size) # Update parameters using their gradient
    train_l=loss(lreg(features,w,b),labels)
    print('epoch %d, loss %f' % (epoch + 1, train_l.mean().asnumpy()))
````

 because we synthesized the data ourselves, we know precisely what the true parameters are. Thus, we can evaluate our success in training by comparing the true parameters with those that we learned through our training loop.

````python
print('Error in estimating w', true_w - w.reshape(true_w.shape))
print('Error in estimating b', true_b - b)
````

# Concise Implementation of Linear Regression
----

````python
import d2l
from mxnet import autograd, gluon, np, npx
npx.set_np()
true_w = np.array([2, -3.4])
true_b = 4.2
features, labels = d2l.synthetic_data(true_w, true_b, 1000)
````

````python
#??gluon.data.ArrayDataset
````

````python
#??gluon.data.DataLoader
````

Rather than rolling our own iterator, we can call upon Gluonʼs data module to read data

````python
# Saved in the d2l package for later use
def load_array(data_arrays, batch_size):
    dataset=gluon.data.ArrayDataset(*data_arrays)
    return gluon.data.DataLoader(dataset,batch_size,shuffle=True)
````

````python

batch_size = 10
data_iter = load_array((features, labels), batch_size)
````

````python
# printing the first batch
for X,y in data_iter:
    print(X,y)
    break
````

# Defining the Model

````python
from mxnet.gluon import nn
net=nn.Sequential()
````

In Gluon, the fully-connected layer is defined in the Dense class. Since we
only want to generate a single scalar output, we set that number to 1

````python
net.add(nn.Dense(1))
````

# Initializing Model Parameters

initializing the parameter (weights and bias) randomly from a normal distribution with mean 0 and standard deviation 0:01.
Both the weight vector and bias will have attached gradients

````python
#??init.Uniform
````

````python
from mxnet import init
net.initialize(init.Normal(sigma=0.01))
````

#  Defining the Loss Function

````python
from mxnet.gluon import loss
````

````python
#help(loss.L2Loss)
````

````python
# # The squared loss is also known as the L2 norm loss
l2_loss=loss.L2Loss()
````

# Defining the Optimization Algorithm

````python
#from mxnet import optimizer
````

````python
#?optimizer.SGD
#gluon.Trainer
````

````python
#  parameters to optimize is from net via net.collect_params())
# optimizer=sgd
trainer = gluon.Trainer(net.collect_params(),'sgd',{'learning_rate': 0.01})
````

# Training

````python
num_epochs=10
for epoch in range(num_epochs+1):
    for X,y in data_iter:
        with autograd.record():
            l2=l2_loss(net(X),y)
        l2.backward()
        trainer.step(batch_size)
    l2=l2_loss(net(features),labels)
    print('epoch %d, loss: %f' % (epoch, l2.mean().asnumpy()))
       
````

````python
w = net[0].weight.data()
print('Error in estimating w', true_w.reshape(w.shape) - w)
b = net[0].bias.data()
print('Error in estimating b', true_b - b)
````

#  Multinomial Logistic Regression (Softmax Classifier)
----

Softmax Classifier often written as $softmax(z)$ can be considered to be a generalization of logistic regression to handle multi-class classification problems. where $y_{i}=1, 2,---,N$ with N being the number of classes. It takes N-dimensional vector of arbitrary real values and produces another N-dimensional vector with real values. Given the number of classes N, the softmax classifier is defined as



$$  \hat{y}=softmax(z)=\begin{bmatrix}
    p(y=1|x;W) \\
     p(y=2|x;W) \\
      p(y=3|x;W) \\
      .\\
      .\\
      .\\
       p(y=N|x;z)  \end{bmatrix} =\frac{}{\sum_{i=0}^{n}\exp(w^{T}x)}  \begin{bmatrix}
    \exp(w_{(1i)}^{T}x_{i}) \\
     \exp(w_{(2i)}^{T}x_{i})\\
     \exp(w_{(3i)}^{T}x_{i}) \\
      .\\
      .\\
      .\\
       \exp(w_{(ni)}^{T}x_{i}) 
  \end{bmatrix} $$
 or
  
  $$ \hat y_{i} =spftmax(z)= \frac{exp(W_{i}^{T}x_{i})}{\sum_{j}W_{j}^{T}x_{i}}  $$
   where $$z=W^{T}X$$ and
   $$w_{1i}x_{i}=w_{11}x_{1} + w_{12}x_{2}+ w_{13}x_{3}$$ and 
   $$w_{2i}x_{i}=w_{21}x_{1} + w_{22}x_{2}+ w_{23}x_{3}$$ in that order
  and with the predicted class being the class with the highest probability
  $$predicted \ class=argmax_{i} \hat y(z) $$
  
  the softmax classifier output values within the range $[0, 1]$. the denominator normalizes the values which causes these class probabilities to sum up to one

# Binary cross entropy

The binary cross-entropy loss (logistic loss) measure the quality of our predicted probabilities by the model and is used in binary classification problems with conditional probability outputs. In binary classification problems, we assume a set of two target classes labeled 0 and 1 in which each imput will be classified as belonging to one of the classes based on it features.
The prediction rule is giving as

$$ \{^{0 \ \ if \ \ \hat y \ \ <0} _{1 \ \ if \ \ \hat y \ \ \geqslant  0}  $$


The binary cross entropy is defined as

$$ \ell(y_{i}, \hat y_{i})=-y_{i}log \hat y_{i} -(1-y_{i})log(1- \hat y_{i})$$
